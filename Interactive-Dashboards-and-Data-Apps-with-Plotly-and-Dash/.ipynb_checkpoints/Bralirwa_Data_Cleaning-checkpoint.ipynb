```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/bralirwa.csv',header=1)
data.head()
```

```python
data_fil = data.drop(['Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8'],axis=1)
```

```python
data_fil.head()
```

```python
#melting our data 
pd.melt(data_fil,id_vars='(In Rwf millions)',
       value_vars=[str(x) for x in range(2017,2021)],
       var_name='year')
```

```python
data_melt = pd.melt(data_fil,id_vars='(In Rwf millions)',
       value_vars=[str(x) for x in range(2017,2021)],
       var_name='year')
data_melt.head()
```

```python
data_melt.duplicated().any()
```

```python
data_mel = data_melt.copy()
data_mel['index'] = data_melt.index
data_mel.head()
```

```python
#pivoting our data to make them look nice
data_piv = data_mel.pivot(index=['index','year'],
               columns='(In Rwf millions)',
               values='value').reset_index()
data_piv.head(4)
```

```python

data_pivot = data_piv.groupby('year').first()
data_pivot
```

```python
trial = data_pivot.copy()
```

```python
trial['Adjustments for:']=np.where(trial['Adjustments for:']==None,'','')
```

```python
trial
```

```python
trial['Cash flow financing activities']=np.where(trial['Cash flow financing activities']==None,'','')
trial['Cash flow from investing activities']=np.where(trial['Cash flow from investing activities']==None,'','')
trial['Cash flow from operating activities']=np.where(trial['Cash flow from operating activities']==None,'','')
trial
```

```python
trial['Changes in working capital']=np.where(trial['Changes in working capital']==None,'','')
trial
```

```python
trial.rename(columns={np.NaN:'Cash Flow from operations(CFO),Total'})
```

```python
trial_rename = trial.rename(columns={np.NaN:'Cash Flow from operations(CFO),Total'})
trial_rename
```

```python
trial_rename = trial_rename.drop('index',axis=1)
```
